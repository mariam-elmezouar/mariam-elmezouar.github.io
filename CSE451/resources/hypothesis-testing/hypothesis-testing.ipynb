# Research Question 1 (RQ1): Does the level of parental education (Medu and Fedu) significantly impact students' final grades (G3)?



### Null Hypothesis (H0): The level of parental education does not significantly impact students' final grades.

### Alternative Hypothesis (H1): The level of parental education significantly impacts students' final grades.

### Test: Multiple Linear Regression



```python
import pandas as pd
import statsmodels.api as sm
from scipy import stats
from scipy.stats import chi2_contingency
```

```python
df = pd.read_csv('student-por.csv')

df.head()
```

```python
# Select the relevant columns for the analysis
data = df[['Medu', 'Fedu', 'G3']]

# Add a constant (intercept) term for the linear regression model
data['intercept'] = 1

# Define the independent variables (parental education) and the dependent variable (final grades)
X = data[['Medu', 'Fedu', 'intercept']]
y = data['G3']

# Fit the multiple linear regression model
model = sm.OLS(y, X).fit()

# Print the summary statistics of the regression model
print(model.summary())
```

## Conclusions

The coefficient for Medu (mother's education) is statistically significant (p < 0.05), indicating that it has a significant impact on students' final grades.

The coefficient for Fedu (father's education) has a p-value of 0.053, which is very close to the 0.05 threshold. While it's not statistically significant at the 5% level, it is still worth considering. Depending on the context and your research objectives, you might choose to interpret it cautiously and investigate further.

Keep in mind that the choice of the significance level (e.g., 0.05, 0.01) can vary depending on the specific requirements of your analysis or research field. If a lower significance level is used (e.g., 0.01), Fedu may not be considered statistically significant.

In summary, the output indicates that the model suggests a statistically significant relationship between maternal education levels (Medu) and students' final grades (G3). However, the R-squared value is relatively low, suggesting that these variables explain only a small portion of the variation in final grades.

Therefore, you would reject the null hypothesis (H0) for Medu and conclude that Medu has a statistically significant impact on students' final grades.

We fail to reject the null hypothesis for Fedu, suggesting that there is not enough evidence to conclude that Fedu has a statistically significant impact on students' final grades. 



# Research Question 2 (RQ2): Is there a significant difference in final grades (G3) between students who receive additional educational support (schoolsup) and those who do not?


### Null Hypothesis (H0): There is no significant difference in final grades between students who receive additional educational support and those who do not.
### Alternative Hypothesis (H1): There is a significant difference in final grades between students who receive additional educational support and those who do not.
### Test: Independent Samples T-Test

```python

# Define the two groups based on the 'schoolsup' variable
group1 = df[df['schoolsup'] == 'yes']['G3']  # Group of students who receive additional support
group2 = df[df['schoolsup'] == 'no']['G3']   # Group of students who do not receive additional support

# Perform an Independent Samples T-Test
t_statistic, p_value = stats.ttest_ind(group1, group2)

# Define the significance level (alpha)
alpha = 0.05

# Print the results and make a significance decision
print("T-Statistic:", t_statistic)
print("P-Value:", p_value)

if p_value < alpha:
    print("Result: Reject the null hypothesis (H0)")
    print("Conclusion: There is a significant difference in final grades between students who receive additional educational support and those who do not.")
else:
    print("Result: Fail to reject the null hypothesis (H0)")
    print("Conclusion: There is no significant difference in final grades between the two groups.")
```

## Conclusions

The calculated t-value is approximately -1.693.
The p-value associated with the t-test is approximately 0.091.
Since the t-value is negative, it suggests that the sample mean of the group receiving additional educational support is slightly smaller than the sample mean of the group not receiving support. However, because p-value is 0.091, we fail to reject the null hypothesis. This indicates that the observed difference in means is not statistically significant, and any observed differences could have occurred by chance.






# Research Question 3 (RQ3): Is there a significant association between students' internet access at home (internet) and their desire to pursue higher education (higher)?


### Null Hypothesis (H0): There is no significant association between students' internet access at home and their desire to pursue higher education.

### Alternative Hypothesis (H1): There is a significant association between students' internet access at home and their desire to pursue higher education.

### Test: Chi-Square Test of Independence

```python

# Create a contingency table by cross-tabulating 'internet' and 'higher'
contingency_table = pd.crosstab(df['internet'], df['higher'])

# Perform the chi-square test of independence
chi2, p_value, _, _ = chi2_contingency(contingency_table)

# Define the significance level (alpha)
alpha = 0.05

# Print the results and make a significance decision
print("Chi-Square Statistic:", chi2)
print("P-Value:", p_value)

if p_value < alpha:
    print("Result: Reject the null hypothesis (H0)")
    print("Conclusion: There is a significant association between students' internet access at home and their desire to pursue higher education.")
else:
    print("Result: Fail to reject the null hypothesis (H0)")
    print("Conclusion: There is no significant association between the two variables.")
```

### Conclusions

#### How to interpret the Chi-Square Statistic:

Magnitude of the Chi-Square Statistic: The larger the Chi-Square Statistic, the greater the deviation between the observed data and what would be expected under the assumption of independence. A larger Chi-Square Statistic indicates stronger evidence against the null hypothesis.

The p-value associated with the chi-square test is 0.1007, which is greater than the chosen significance level (alpha = 0.05). When the p-value is greater than alpha, it indicates that we fail to reject the null hypothesis (H0).

# A summary of statistical tests and when to use each


| Statistical Test        | When to Use It               | Dataset Attributes                                            |
|-------------------------|------------------------------|--------------------------------------------------------------|
| t-Test (Independent Samples) | Compare means of two groups | - Continuous dependent variable<br> - Categorical independent variable (2 groups) |
| Paired t-Test           | Compare means of paired data | - Continuous dependent variable<br> - Paired or repeated measurements |
| Chi-Square Test of Independence | Assess association between categorical variables | - Categorical variables in a contingency table                |
| One-Way ANOVA           | Compare means of more than two groups | - Continuous dependent variable<br> - Categorical independent variable (with >2 levels) |
| Two-Way ANOVA           | Assess interaction effects between two categorical variables | - Continuous dependent variable<br> - Two categorical independent variables |
| Pearson's Correlation   | Examine linear relationship between two continuous variables | - Two continuous variables                                   |
| Spearman's Rank Correlation | Assess monotonic relationship between two variables | - Ordinal or continuous variables                            |
| Mann-Whitney U Test     | Compare distributions of two groups (non-parametric) | - Continuous dependent variable<br> - Categorical independent variable (2 groups) |
| Kruskal-Wallis Test     | Compare distributions of more than two groups (non-parametric) | - Continuous dependent variable<br> - Categorical independent variable (with >2 levels) |
| Linear Regression       | Predict a continuous outcome based on predictor variables | - Continuous dependent variable<br> - One or more predictor variables |
| Logistic Regression     | Predict a binary outcome based on predictor variables | - Binary dependent variable (0/1)<br> - One or more predictor variables |
